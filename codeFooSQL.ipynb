```python
conn = sqlite3.connect('codeFoo.db')
cursor = conn.cursor()
create_media = """
CREATE TABLE IF NOT EXISTS media (
    id BIGINT,
    media_type TEXT,
    name TEXT,
    short_name TEXT,
    long_description TEXT,
    short_description TEXT,
    created_at TEXT,
    updated_at TEXT,
    review_url TEXT,
    review_score DOUBLE,
    slug TEXT,
    genres TEXT,
    created_by TEXT,
    published_by TEXT,
    franchises TEXT,
    regions TEXT,
    PRIMARY KEY (id)
)
"""
cursor.execute(create_media)
```

```python
insert_data = """
INSERT INTO media (
    id,
    media_type,
    name,
    short_name,
    long_description,
    short_description,
    created_at,
    updated_at,
    review_url,
    review_score,
    slug,
    genres,
    created_by,
    published_by,
    franchises,
    regions)
VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
"""
cursor.executemany(insert_data, csvreader)
```

```python
select_all = "SELECT * FROM media"
rows = cursor.execute(select_all).fetchall()
for r in rows:
    print(r)
```
